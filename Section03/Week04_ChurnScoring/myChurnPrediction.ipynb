# STEP 1
- CLEAN DATA

````python
df['SHOP_DATE'] = pd.to_datetime(df['SHOP_DATE'], format='%Y%m%d')

# drop NaN (for non-member only)

filnanfordf = ['CUST_PRICE_SENSITIVITY','CUST_LIFESTAGE']

for x in filnanfordf :
    df[x] = df[x].fillna('XX')

df.dropna(inplace=True)

df.head()
````

- Prepare DATA (Transform)

````python
df['order_month'] = df['SHOP_DATE'].dt.to_period('M')

df.loc[:,['SHOP_DATE','order_month']].head()
````

````python
from operator import attrgetter

df['start_month'] = df.groupby('CUST_CODE')['SHOP_DATE'].transform('min').dt.to_period('M')
df['period_number'] = (df['order_month'] - df['start_month']).apply(attrgetter('n')).astype(int)

print(f'#### SIMPLE ####\n')
display(df.loc[:,['CUST_CODE','SHOP_DATE','order_month','start_month','period_number']])

# Preparing for a PLOT

print(f'#### FOR COHORT ####\n')
df_cohort = df.groupby(['start_month', 'order_month']).agg(n_customers=('CUST_CODE', 'nunique')).reset_index(drop=False)
df_cohort['period_number'] = (df_cohort['order_month'] - df_cohort['start_month']).apply(attrgetter('n'))

cohort_pivot = df_cohort.pivot_table(index='start_month', columns='period_number', values='n_customers')

display(cohort_pivot)
````

- PLOT : Cohort Analysis

````python
# Separate column 0 (number of new customer in each months)
cohort_size = cohort_pivot.iloc[:, 0]
cohort_size = pd.DataFrame(cohort_size).rename(columns={0: 'new_member_size'})

# Divided by number of start months
cohort_pivot = cohort_pivot.divide(cohort_size['new_member_size'], axis=0)
````

````python
# PLOT

import matplotlib.colors as mcolors

sns.axes_style("white")
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 14), sharey=True, gridspec_kw={'width_ratios': [1, 11]})

# retention matrix
sns.heatmap(cohort_pivot, 
            mask=cohort_pivot.isnull(), 
            annot=True, 
            fmt='.0%', 
            cmap='viridis', 
            ax=ax[1])
ax[1].set_title('Monthly Cohorts: User Retention', fontsize=20)
ax[1].set(xlabel='#number of periods',
            ylabel='')

# cohort size
white_cmap = mcolors.ListedColormap(['white'])
sns.heatmap(cohort_size, 
            annot=True, 
            cbar=False, 
            fmt='g', 
            cmap=white_cmap, 
            ax=ax[0])

fig.tight_layout()
````

- Assigning STATUS

````python
churn = { 'CUST_CODE':[] ,'churn_score':[] }

for i in newdf['CUST_CODE'].unique() :

    x = newdf.loc[ (newdf['CUST_CODE'] == i) ]

    startmonth = newdf.loc[ (newdf['CUST_CODE'] == i) ]['start_month'].iloc[0]
    maxpr = newdf.loc[ (newdf['CUST_CODE'] == i) , ]['max_period_number'].unique()[0]
    active_count = 0
    
    for i0,j in enumerate(x['period_number']):
        if i0==0 and i0==j : 
            newdf.loc[ (newdf['CUST_CODE'] == i) ,['period_number'] ] = newdf.loc[ (newdf['CUST_CODE'] == i) 
                                                                                  ,['period_number'] ].replace(j,'NEW')
            n=j
            active_count += 1

        elif n+1==j : 
            newdf.loc[ (newdf['CUST_CODE'] == i) ,['period_number'] ] = newdf.loc[ (newdf['CUST_CODE'] == i) 
                                                                                  ,['period_number'] ].replace(j,'REPEAT')
            n+=1
            active_count += 1

        elif ( n+1 < j ) :
            newdf.loc[ (newdf['CUST_CODE'] == i) ,['period_number'] ] = newdf.loc[ (newdf['CUST_CODE'] == i) 
                                                                                  ,['period_number'] ].replace(j,'REACTIVATE')
            # add CHURN label
            for i1,k in enumerate(range(n+1,j)):
                temp = newdf.loc[ (newdf['period_number'].isin(['NEW','REPEAT','REACTIVATE'])) 
                                 & (newdf['CUST_CODE'] == i) 
                                ]['order_month'].iloc[-2] + pd.DateOffset(months=i1+1)
        
                Xtemp = { 
                    'CUST_CODE' : i ,
                    'order_month' : temp, 
                    'start_month' : startmonth,
                    'period_number' : 'CHURN',
                    'max_period_number' : maxpr
                            }
                newdf = pd.concat([newdf, pd.DataFrame(Xtemp, index=[0])], ignore_index = True)

            n=j
            active_count += 1
        
    # add CHURN label round 2
    for i1,k in enumerate(range(n+1,maxpr+1)):
        temp = newdf.loc[ (newdf['period_number'].isin(['NEW','REPEAT','REACTIVATE'])) 
                         & (newdf['CUST_CODE'] == i) 
                            ]['order_month'].iloc[-1] + pd.DateOffset(months=i1+1)

        Xtemp = { 
            'CUST_CODE' : i ,
            'order_month' : temp, 
            'start_month' : startmonth,
            'period_number' : 'CHURN',
            'max_period_number' : maxpr
                    }
        newdf = pd.concat([newdf, pd.DataFrame(Xtemp, index=[0])], ignore_index = True)

    churn_score = (maxpr-active_count)/maxpr
    churn['CUST_CODE'].append(i)
    churn['churn_score'].append(churn_score)
````

- Assigning CHURN Score <br>

```CHURN Score = [(number of months since registeration)-(active_count)]/(number of months since registeration)```


````python
churn = pd.DataFrame.from_dict(churn)

# assign churn score by customer id (one-to-many)

score_to_id_mapper = churn.set_index('CUST_CODE')['churn_score']
newdf['churn_score'] = newdf['CUST_CODE'].map(score_to_id_mapper)
````
